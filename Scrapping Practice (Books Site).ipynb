```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

PATH = "C:\chromedriver.exe"

driver=webdriver.Chrome(PATH)

driver.implicitly_wait(10)
driver.get("http://books.toscrape.com/catalogue/birdsong-a-story-in-pictures_975/index.html")

driver.set_window_size(1024, 600)
driver.maximize_window()

table=[]
table.append(['Product Type','Book Name','Genre','UPC','Price (excl. tax)','Price (incl. tax)','Tax','Availability','Number of Reviews'])
row=[]
ProdType=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[2]/td').text.replace(",","`").replace("\'","~")
BookName=driver.find_element_by_tag_name("h1").text.replace(",","`").replace("\'","~")
Genre=driver.find_element_by_xpath('//*[@id="default"]/div/div/ul/li[3]/a').text.replace(",","`").replace("\'","~")
UPC=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[1]/td').text.replace(",","`").replace("\'","~")
PriceExTax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[3]/td').text.replace(",","`").replace("\'","~")
PriceInTax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[4]/td').text.replace(",","`").replace("\'","~")
Tax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[5]/td').text.replace(",","`").replace("\'","~")
Avail=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[6]/td').text.replace(",","`").replace("\'","~")
NumRev=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[7]/td').text.replace(",","`").replace("\'","~")
row.append(ProdType)
row.append(BookName)
row.append(Genre)
row.append(UPC)
row.append(PriceExTax)
row.append(PriceInTax)
row.append(Tax)
row.append(Avail)
row.append(NumRev)
table.append(row)
print(table)

f=open('f1.txt','w')
for item in table:
    f.write("%s\n"%str(item).replace("[","").replace("]","").replace("\"","").replace("\'","").replace(",",";").replace("`",",").replace("~","\'"))
f.close()
```

```python
%%time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import time


caps = DesiredCapabilities().CHROME
caps["pageLoadStrategy"] = "none"

PATH = "C:\chromedriver.exe"

driver=webdriver.Chrome(PATH)

mainTab=driver.current_window_handle

table=[]
table.append(['Product Type','Book Name','Genre','UPC','Price (excl. tax)','Price (incl. tax)','Tax','Availability','Number of Reviews'])

secTabs=[i for i in range(20)]
tabCount=0

for count in range(20):
    driver.execute_script('''window.open("_blank","_blank");''')

for windowHandle in driver.window_handles:
        if windowHandle!=mainTab:
            secTabs[tabCount]=windowHandle
            tabCount+=1    

driver.switch_to.window(mainTab)
driver.get('http://books.toscrape.com/index.html')

while True:
    while len(driver.find_elements_by_class_name("image_container"))!=20:
        time.sleep(0.1)

    links=[]
    aElements=driver.find_elements_by_class_name("image_container")
    for aElement in aElements:
        aElement=aElement.find_element_by_tag_name("a")
        links.append(aElement.get_attribute("href"))

    for x in range(20):
        driver.switch_to.window(secTabs[x])
        driver.get(links[x])

    driver.switch_to.window(secTabs[0])
    
    for x in range(20):
        driver.switch_to.window(secTabs[x])
        while len(driver.find_elements_by_tag_name("img"))==0:
            time.sleep(0.1)
        row=[]
        ProdType=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[2]/td').text.replace(",","`").replace("\'","~")
        BookName=driver.find_element_by_tag_name("h1").text.replace(",","`").replace("\'","~")
        Genre=driver.find_element_by_xpath('//*[@id="default"]/div/div/ul/li[3]/a').text.replace(",","`").replace("\'","~")
        UPC=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[1]/td').text.replace(",","`").replace("\'","~")
        PriceExTax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[3]/td').text.replace(",","`").replace("\'","~")
        PriceInTax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[4]/td').text.replace(",","`").replace("\'","~")
        Tax=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[5]/td').text.replace(",","`").replace("\'","~")
        Avail=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[6]/td').text.replace(",","`").replace("\'","~")
        NumRev=driver.find_element_by_xpath('//*[@id="content_inner"]/article/table/tbody/tr[7]/td').text.replace(",","`").replace("\'","~")
        row.append(ProdType)
        row.append(BookName)
        row.append(Genre)
        row.append(UPC)
        row.append(PriceExTax)
        row.append(PriceInTax)
        row.append(Tax)
        row.append(Avail)
        row.append(NumRev)
        table.append(row)
    
    driver.switch_to.window(mainTab)
    if driver.find_elements_by_class_name("next")==[]:
        break;
    else:
        driver.find_element_by_class_name("next").find_element_by_tag_name("a").click()

f=open('f1.txt','w')
for item in table:
    f.write("%s\n"%str(item).replace("[","").replace("]","").replace("\"","").replace("\'","").replace(",",";").replace("`",",").replace("~","\'"))
f.close()

driver.quit()
```
